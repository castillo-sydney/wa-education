```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
education = gpd.read_file('data/No_High_School_Diploma')
education = education.dropna()
education.head()
```

```python
employment = gpd.read_file('data/Unemployed_Population')
employment.head()
```

```python
education['Percent_No_Diploma'] = education['Num_Withou'] / education['Population'] * 100
education = education.drop('Percent_Wi', 1)
education.head()
```

```python
employment['Percent_Unem'] = employment['Num_Unempl'] / employment['Employable'] * 100
employment = employment.drop('Percent_Un', 1)
employment.head()
```

```python
fig, ax = plt.subplots(1)
education.plot(ax=ax, color='#EEEEEE')
education.plot(ax=ax, column='Percent_No_Diploma', legend=True)
```

```python
fig, ax = plt.subplots(1)
employment.plot(ax=ax, color='#EEEEEE')
employment.plot(ax=ax, column='Percent_Unem', legend=True)
```
